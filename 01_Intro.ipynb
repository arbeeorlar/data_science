```python
subset = asian_cars[["vehicle_weight", "model_year"]]
```

```python
subset7 = subset.head(7)
```

```python
x = subset7.values
```

```python
XTX = x.T @ x
```

```python
XTX_inv = np.linalg.inv(XTX)
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
y = y[:x.shape[0]]
w = XTX_inv @ x.T @ y
```

```python
w_sum = w.sum()
```
